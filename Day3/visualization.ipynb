```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y,marker = 'o', color = 'r', linestyle = '--')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph')
plt.grid(True)
plt.show()
```

```python
plt.scatter(x,y,marker = '*', color = 'r')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Graph')
plt.show()
```

```python
plt.bar(x,y,color = 'c')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Graph')
plt.show()
```

```python
import numpy as np

data = np.random.randn(1000)
plt.hist(data, bins=30, color = 'beige', edgecolor = 'black')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
sizes = [20, 30, 40, 50]
colors = ['red', 'blue', 'green', 'yellow']
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, colors=colors, labels=labels, autopct='%1.1f%%', shadow=True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 

sales = np.array([100, 120, 130, 140, 150,170,230,240,250,270])
days = np.arange(1, 11)

print(sales)

total_sales = np.sum(sales)
average_sales = np.mean(sales)
highest_sales = np.max(sales)
lowest_sales = np.min(sales)

print("Total Sales: ", total_sales)
print("Average Sales: ", average_sales)
print("Highest Sales: ", highest_sales)
print("Lowest Sales: ", lowest_sales)

plt.figure(figsize=(8,5))
plt.plot(days, sales, marker = 'o', linestyle = '-' , color = 'b', label = 'sales')

plt.axhline(average_sales, color='r', linestyle='--', label = 'Average Sales')

plt.xlabel('Days')
plt.ylabel('Sales Revenue')
plt.title('Sales Trend over years')
plt.legend()
# plt.grid(True)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)
plt.show()
```

```python
plt.figure(figsize=(6, 6,))
plt.pie(sales, labels = days, colors = plt.cm.Paired.colors, autopct = '%1.1f%%')
plt.title('Sales Distribution')
plt.show()
```
